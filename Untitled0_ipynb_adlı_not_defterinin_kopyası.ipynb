```python
class QAEngine:
    def __init__(self, api_key):
        self.api_key = api_key
        self.api_url = f"https://generativelanguage.googleapis.com/v1beta/models/gemini-1.5-flash:generateContent?key={api_key}"

    def get_answer(self, context, question):
        headers = {
            "Content-Type": "application/json"
        }
        data = {
            "contents": [
                {
                    "parts": [
                        {"text": f"Context: {context}\nQuestion: {question}\nAnswer:"}
                    ]
                }
            ]
        }
        response = requests.post(self.api_url, headers=headers, data=json.dumps(data))
        if response.status_code == 200:
            return response.json()['candidates'][0]['content']['parts'][0]['text']
        else:
            raise Exception(f"Error: {response.status_code}, {response.text}")
```

```python
# QAEngine nesnesini oluşturma
qa_engine = QAEngine(API_KEY)

# Context ve soru tanımlıyoruz
context = "Elazığ'ın havası hem kuru hem sıcaktır. Rozacea hastalığına iyi gelmez."
question = "Elazığ havası nasıldır?"

# Cevabı alıyoruz
try:
    answer = qa_engine.get_answer(context, question)
    print("Cevap:", answer)
except Exception as e:
    print("Hata:", e)
```

```python
from google.colab import files
import fitz  # PyMuPDF (PDF okuma için)
from docx import Document
from IPython.display import display, HTML

# Dosyayı yükleme
uploaded = files.upload()

# Yüklenen dosyanın adını alıyoruz
filename = list(uploaded.keys())[0]

# .docx dosyasını okuma fonksiyonu
def read_docx(file_path):
    doc = Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# .pdf dosyasını okuma fonksiyonu
def read_pdf(file_path):
    doc = fitz.open(file_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text

# Dosya türüne göre içerik okuma
if filename.endswith('.docx'):
    context = read_docx(filename)
elif filename.endswith('.pdf'):
    context = read_pdf(filename)
else:
    context = "Desteklenmeyen dosya türü."

# İlk 500 karakteri yazdırıyoruz
print("Context (Dosya İçeriği):")
print(context[:500])

# Uzun cevabı alt satıra geçirecek şekilde HTML kullanma
def wrap_text_for_display(text):
    wrapped_text = "<pre style='white-space: pre-wrap; word-wrap: break-word;'>"
    wrapped_text += text
    wrapped_text += "</pre>"
    display(HTML(wrapped_text))

# Cevabın uzun olması durumunda alt satıra geçmesini sağlamak
wrap_text_for_display(context)

# Soru almak
question = input("\nSormak istediğiniz soru: ")

# QAEngine nesnesini kullanarak cevabı alıyoruz
try:
    answer = qa_engine.get_answer(context, question)
    print("\n--- Cevap ---")
    wrap_text_for_display(answer)  # Cevabı da alt satıra geçiriyoruz
except Exception as e:
    print("Hata:", e)
```
